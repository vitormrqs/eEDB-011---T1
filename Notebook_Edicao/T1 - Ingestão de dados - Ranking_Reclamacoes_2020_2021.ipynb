# Junção dos 4 csv's

```python
  # using merge function by setting how='inner'
Ranking_Reclamacoes_2020_2021 = pd.concat([data1, data2, data3, data4, data8, data7, data6, data5])
Ranking_Reclamacoes_2020_2021 = Ranking_Reclamacoes_2020_2021[Ranking_Reclamacoes_2020_2021['CNPJ IF'] != ' ']
Ranking_Reclamacoes_2020_2021 = Ranking_Reclamacoes_2020_2021.iloc[:,:-1]
Ranking_Reclamacoes_2020_2021 = Ranking_Reclamacoes_2020_2021.reset_index(drop=True)
```

```python
CNPJ = 58616418
url = "https://olinda.bcb.gov.br/olinda/servico/Informes_ListaTarifasPorInstituicaoFinanceira/versao/v1/odata/ListaTarifasPorInstituicaoFinanceira(PessoaFisicaOuJuridica=@PessoaFisicaOuJuridica,CNPJ=@CNPJ)?@PessoaFisicaOuJuridica='J'&@CNPJ='"+str(CNPJ)+"'&$top=100&$format=json"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
json = response.json()

#print(response.text)
print(len(json['value']))
```

```python
#Loop ValorMaximo
ValorMaximo = 0 

for j in json['value']:
    ValorMaximo += j['ValorMaximo']

print(ValorMaximo)
```
